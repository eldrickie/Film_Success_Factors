```python
basic_df = pd.read_csv('imdb.title.basics.csv.gz')
gross_df = pd.read_csv('bom.movie_gross.csv.gz')
ratings_df = pd.read_csv('imdb.title.ratings.csv.gz')
budget_gross = pd.read_csv('tn.movie_budgets.csv.gz')
```

#### Merges and Data Cleaning

```python
basic_df = basic_df.set_index('tconst')
ratings_df = ratings_df.set_index('tconst')
basicRatings = basic_df.join(ratings_df)
# Because I will be joining this dataframe first through column 'tconst' and subsequently, 'primary_title'; 
# I will address the null values after the merges
```

```python
basicRatings = basicRatings[pd.notnull(basicRatings['genres'])]
# Because the same films that are missing genres, are also missing the majority of other columns,
#I dropped the rows which contain null values for the column 'genres'.
```

```python
genres = [genre.split(',') for genre in basicRatings['genres']]
genre_count = []
for element in genres:
    for count in element:
        if count not in genre_count:
            genre_count.append(count)
for count in genre_count:
    basicRatings[count] = 0
basicRatings.reset_index(inplace = True)
```

```python
for index, genre in enumerate(basicRatings['genres']):
    for count in genre_count:
        if count in genre:
            basicRatings.at[index, count] = 1
```

```python
fig, ax = plt.subplots(figsize = (30,10))
ax.bar(x = genre_count , height = basicRatings[genre_count].sum(), color = 'pink')
ax.set_xlabel('Genres of Films on IMBDB', fontsize = 15)
ax.set_ylabel('Number of Films Belonging to Genre', fontsize = 15)
ax.set_title('Occurences of Films by Genre', fontsize = 20);
```

```python
for punc in punctuation:
    basicRatings['primary_title'] = basicRatings['primary_title'].str.replace(punc, '')
    gross_df['title'] = gross_df['title'].str.replace(punc, '')
basicRatings['primary_title'] = basicRatings['primary_title'].str.lower()
gross_df['title'] = gross_df['title'].str.lower()
```

```python
gbr_df = basicRatings.merge(gross_df, how = 'left', left_on = 'primary_title', right_on = 'title')
```

```python
gbr_df = gbr_df.loc[gbr_df['domestic_gross'].isna() == False]
```

```python
tb_removed = ['Adult', 'Game-Show', 'Short', 'Reality-TV', 'Talk-Show']
for genre in tb_removed:
    genre_count.remove(genre)
gbr_df.drop(tb_removed, axis='columns', inplace = True) 
# My analyses will ultimately use domestic_gross, foreign_gross, and total_gross as a measure of success.
# Dropping rows which contain null values for domestic_gross removed films classified as 'Adult', 
# 'Talk-Show', 'Reality-TV', 'Game-Show', or 'Short'. Thus, I will be dropping these columns.
```

```python
gbr_df['foreign_gross'] = gbr_df['foreign_gross'].str.replace(',', "").astype(float)
# The column 'foreign_gross' in gbr_df is filled with strings instead interger or float values. 
# I will be converting the strings to float values in order to mirror the data type used in the 
# 'domestic_gross' column.
```

### Domestic and Foreign Gross Analysis

In order to keep the rows for column 'foreign_gross': I calculated the mean for the rows in 'domestic_gross' which does not have null value for 'foreign_gross', calculated the mean for the 'foreign_gross' non-null values, created a variable called ratio which contains the ratio between the two means, and applied that ratio on 'domestic_gross' in rows where 'foreign_gross' is null. This process is framed by the the mathematical concepts of Ratios of Means and Ratio Estimation. This process allowed me to keep the rows, by plugging in predictive numbers for null 'foreign_gross' cells.

```python
meanDom = gbr_df.loc[gbr_df['foreign_gross'].isna() == False]['domestic_gross'].mean()
meanForeign = gbr_df['foreign_gross'].mean()
ratio = (meanForeign / meanDom)
gbr_df['foreign_gross'].fillna(round(gbr_df['domestic_gross'] * ratio), inplace = True)
```

### Gross and Genre

```python
genreDom = [(genre, gbr_df.loc[gbr_df[genre] == 1]
             ['domestic_gross'].sum()) for genre in genre_count]
genreForeign = [(genre, gbr_df.loc[gbr_df[genre] == 1]
             ['foreign_gross'].sum()) for genre in genre_count]
```

```python
fig, ax = plt.subplots(figsize = (30,10))
ax.bar(x = [i[0] for i in genreDom] , height = [i[1] for i in genreDom])
ax.set_xlabel('Genres', fontsize = 20)
ax.set_ylabel('Domestic Box-office Earnings in Tens of Billions $', fontsize = 20)
ax.set_title('Domestic Gross Success Grouped by Genre', fontsize = 25);
```

```python
fig, ax = plt.subplots(figsize = (30,10))
ax.bar(x = [i[0] for i in genreForeign] , height = [i[1] for i in genreForeign])
ax.set_xlabel('Genres', fontsize = 20)
ax.set_ylabel('Foreign Box-office Earnings in Tens of Billions $', fontsize = 20)
ax.set_title('Foreign Gross Success Grouped by Genre', fontsize = 25);
```

#### Earnings Per Film Analysis Grouped by Genre
Which genre earns the most per film?

```python
genreDom_mean = [(genre, gbr_df.loc[gbr_df[genre] == 1]
                  ['domestic_gross'].mean()) for genre in genre_count]
genreForeign_mean = [(genre, gbr_df.loc[gbr_df[genre] == 1]
                      ['foreign_gross'].mean()) for genre in genre_count]
```

#### Merging Production Budget Dataset

```python
for punc in punctuation:
    budget_gross['movie'] = budget_gross['movie'].str.replace(punc, '')
budget_gross['movie'] = budget_gross['movie'].str.lower()
# I removed the punctuation in the column 'movie' and converted the result 
# to lowercase in order to merge it with the preceding dataframe
```

```python
budge_gbr = gbr_df.merge(
    budget_gross, how = 'left', left_on = 'primary_title', right_on = 'movie'
)
```

```python
budge_gbr.drop(columns = ['domestic_gross_y', 'worldwide_gross', 'year', 'runtime_minutes',
                          'averagerating', 'numvotes', 'movie', 'id', 'studio'], axis = 1, inplace = True)
# These columns were dropped in order to clean up the dataframe.
# The columns are not used in the analysis
budge_gbr = budge_gbr[pd.notnull(budge_gbr['production_budget'])]
for punc in punctuation:
    budge_gbr['production_budget'] = budge_gbr['production_budget'].str.replace(punc, '')
budge_gbr['production_budget'] = budge_gbr['production_budget'].astype('float')
budge_gbr['total_gross'] = budge_gbr['domestic_gross_x'] + budge_gbr['foreign_gross']
```

### Gross vs. Budget by Genre

```python
totalGross_mean = [(genre, budge_gbr.loc[budge_gbr[genre] == 1]
             ['total_gross'].mean()) for genre in genre_count]
budget_mean = [(genre, budge_gbr.loc[budge_gbr[genre] == 1]
             ['production_budget'].mean()) for genre in genre_count]
```

```python
budgetGrossMean_Ratio = [(genre, (budge_gbr.loc[budge_gbr[genre] == 1]
                        ['total_gross'].mean()/ budge_gbr.loc[budge_gbr[genre] == 1]
                        ['production_budget'].mean())) for genre in genre_count]
# I took the mean of the column 'total_gross' categorized by genre and divided it by the mean
# of the column 'production_budget', also categorized by genre.
```

#### Boxoffice Success in Comparison to Production Budget
Which genre earns the most per film in comparison to its budget?

```python
labels = [i[0] for i in budget_mean]
budge = [i[1] for i in budget_mean]
totals = [i[1] for i in totalGross_mean]

x = np.arange(len(labels))
width = 0.40

fig, ax = plt.subplots(figsize = (30,10))
ax.bar(x-width/2, budge, width, label = 'Budget')
ax.bar(x+width/2, totals, width, label = 'Worldwide Gross')
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize =12)
ax.legend(fontsize = 'xx-large' )
ax.set_xlabel('Genres', fontsize = 18)
ax.set_ylabel('per $ 100 Million', fontsize = 18)
ax.set_title('Budget Compared to Worldwide Gross by Genre)', fontsize= 24);
```

```python
fig, ax = plt.subplots(figsize = (30,10))
ax.bar(x = [i[0] for i in budgetGrossMean_Ratio], 
       height = [i[1] for i in budgetGrossMean_Ratio], 
       color = 'purple')
ax.set_xlabel('Genres', fontsize = 20)
ax.set_ylabel('Ratio', fontsize = 20)
ax.set_title('Worldwide Gross to Production Budget Ratio', fontsize = 25);
```

### Gross by Month

```python
budge_gbr['release_date']= pd.to_datetime(budge_gbr['release_date'])
months = []
for date in budge_gbr['release_date']:
    months.append(date.strftime("%b"))
budge_gbr['Month'] = months
```

```python
months_dict = {
    "Jan":0, "Feb":1, "Mar":2, "Apr":3, "May":4, "Jun":5,
    "Jul":6, "Aug":7, "Sep":8, "Oct":9, "Nov":10, "Dec":11 
}
means = budge_gbr.groupby('Month').mean()['total_gross']
indices = means.index
months_means_dict = dict(zip(indices, means))
for key, value in months_means_dict.items():
    months_dict[key] = value
```

```python
sums = budge_gbr.groupby('Month').sum()['total_gross']
fig, ax = plt.subplots(figsize = (40,20))
sums.plot(ax=ax, kind = 'barh', color = 'blue')
ax.set_ylabel('Months', fontsize = 30)
ax.set_xlabel('Overall Worlwide Earnings in $ Hundreds of Millions', fontsize = 30)
ax.set_title('Total Worldwide Boxoffice Success by Month', fontsize = 35)
plt.xticks(fontsize= 20)
plt.yticks(fontsize= 20);
```

#### Average Earnings Per Film by Genre
Which month of release will allow Microsoft to maximize earnings?

```python
fig, ax = plt.subplots(figsize = (40,20))
sns.barplot(x = [key for key, value in months_dict.items()] , y = [value for key, value in months_dict.items()])
ax.set_ylabel('Mean Worlwide Earnings in $ Hundreds of Millions', fontsize = 30)
ax.set_xlabel('Months', fontsize = 30)
ax.set_title('Mean Worldwide Boxoffice Success by Month', fontsize = 35)
plt.xticks(fontsize= 20)
plt.yticks(fontsize= 20);
```

## Takeaways for Microsoft

- The top three performing genres in both domestic and foreign boxoffices in order were sci-fi, adventure, and animation.
    Microsoft should focus on these three genres if they want to release films, which on average, produce more earnings than other genres. 
- The earnings to production budget ratio suggests that music, sport, then horror films make more money in relation to how much the studio spends in creating each film. However, animation and sci-fi provide comparable ratios while maintaing high mean earnings; The studio may earn similarly with animation and sci-fi with respect to ratio, while earning more per film than music, sport, or horror.  This analysis asserts that these genres provide studios 'more bang for their buck'
- The highest earning films on average were released during the months of June and November. This suggests that releasing films in either month may contribute to the success in boxoffice numbers. For example, films released in September earned half as much as films released in either June or November

### Conclusion
---------------------------------------
Using these metrics, will allow Microsoft to predictively gauge the successes of future films while allowing for flexibility. For example, the new studio may choose to release films which focus on maximizing individual movie earnings by shooting an adventure film and aim to release it in June. Another tactic the studio can implement is to release music or sport themed films. Releasing a large volume of such films compounds good profits in relation to production costs. The studio may choose to stagger the releases between the summer months and the holiday season in order to cast a wider net.
